```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
#Using Logistic Regression as the model

y_overall = iris.iloc[:,-1].values
x_overall= iris.iloc[:,:-1].values
print('The shape of input vector is: ',x_overall.shape)
print ('The shape of output vector is: ', y_overall.shape)
#Define the input and output vectors to interact with the model
```

```python
x_train,x_test, y_train, y_test = train_test_split(x_overall, y_overall, test_size=0.1)
print('The shape of training input vector is: ', x_train.shape)
print ('The shape of testing input vector is: ', x_test.shape)
#Splitting the model for testing and training in the ratio 90% to 10%
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
#Train, i.e, fit the model to the training set
```

```python
prediction_output = model.predict(x_test)
print (prediction_output)
print (y_test)
#Use the model for prediction
```

```python
loss = []
for i in range (len(prediction_output)):
    if (prediction_output[i]==y_test[i]):
        loss.append(0)
    else:
        loss.append(1)
print ('The error percentage is: ',np.average(loss)*100,"%")
```

```python
number_name = { 1:'Iris-setosa', 2:'Iris-versicolor', 3:'Iris-virginica'}
prediction_names =np.array( [number_name[item] for item in prediction_output]).reshape(len(y_test),1)
y_test = np.array([number_name[item] for item in prediction_output]).reshape(len(y_test),1)
#Assign name values to previously given numeric value for each species
```

```python
pred_act = np.concatenate((prediction_names, y_test), axis = 1)
pred_act[:5]
#Put the predicted and actual values side by side column wise
```

```python
pd.DataFrame(data = pred_act, columns = ['Predictions', 'Actual'],).to_csv('Predictions_iris.csv',  index = False)
#Convert the result into dataframe
#Convert the dataframe into a csv file called 'Predictions.csv'
```
